```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
world_pop = pd.read_excel("world-population.xlsm")
```

```python
world_pop
```

### Step Chart

```python
plt.step(world_pop.Year, world_pop.Population)
plt.xlabel("Year"), plt.ylabel("Population (Billions)")
plt.title("World Population by Year")
plt.show()
```

### Line Chart

```python
plt.plot(world_pop.Year, world_pop.Population)
plt.xlabel("Year"), plt.ylabel("Population (Billions)")
plt.title("World Population by Year")
plt.show()
```
