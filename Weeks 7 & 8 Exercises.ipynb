## Scatterplot

```python
# filter out DC
filtered = crime.loc[crime["state"] != "District of Columbia"]
```

```python
sns.scatterplot(data = filtered, x = "murder", y='aggravated_assault' )
plt.xlabel("Murder per 100,000"), plt.ylabel("Aggravated Assault per 100,000")
plt.title("Murder vs Aggravated Assault in the U.S.")
plt.show()
```

## Bubble Chart

```python
filtered2 = filtered.loc[crime["state"] != "United States"]
```

```python
sns.scatterplot(data=filtered2, x="murder", y="aggravated_assault", size="population", legend=True, sizes=(20, 1000))
plt.xlabel("Murder per 100,000"), plt.ylabel("Aggravated Assault per 100,000")
plt.title("Murder vs Aggravated Assault in the U.S.")
```

## Density Plot

```python
sns.kdeplot(birth_rate["1980"])
plt.xlabel("Births per 1,000 population"), plt.title("Global Distribution of Birth Rates, 1980")
plt.show()
```
