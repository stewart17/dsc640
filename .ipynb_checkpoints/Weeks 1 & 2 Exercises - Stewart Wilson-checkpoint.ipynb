```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read hotdog data
hotdog = pd.read_excel("hotdog-contest-winners.xlsm")
hotdog
```

```python
# read hotdog places data
hd_places = pd.read_excel("hotdog-places.xlsm")
hd_places
```

```python
# read obama data
obama = pd.read_excel("obama-approval-ratings.xls")
obama
```

```python
plt.bar(hotdog["Year"], hotdog["Dogs eaten"])
plt.xlabel("Year")
plt.ylabel("Hot Dogs Eaten")
plt.title("Number of Hot Dogs Eaten by Year")
plt.show()
```

```python
plt.pie(hotdog.groupby("Country")["Dogs eaten"].sum(), labels = ["Germany", "Japan", "Mexico", "United States"])
plt.title("Number of Dogs Eaten by Country")
plt.show()
```

```python
explode = (0.05, 0.05, 0.05, 0.05)
plt.pie(hotdog.groupby("Country")["Dogs eaten"].sum(), 
        labels = ["Germany", "Japan", "Mexico", "United States"], autopct='%1.1f%%', 
        pctdistance = .85, explode = explode)
plt.title("Number of Dogs Eaten by Country")

centre_circle = plt.Circle((0,0), .5, fc="white")
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

```python
plt.bar(obama.Issue, obama.Approve, color = 'r')
plt.bar(obama.Issue, obama.Disapprove, bottom = obama.Approve, color = "b")
plt.bar(obama.Issue, obama["None"], bottom = obama.Approve + obama.Disapprove, color="g")
plt.xticks(rotation = 60)
plt.legend(["Approve", "Disapprove", "None"])
plt.title("Approval Rating by Issue")
plt.show()
```
