```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import squarify
```

```python
unemployment = pd.read_csv("unemployement-rate-1948-2010.csv")
expenditures = pd.read_csv("expenditures.txt", sep="\t")
```

```python
unemployment
```

```python
expenditures
```

## Area Chart

```python
plt.fill_between(unemployment.Year, unemployment.Value)
plt.xlabel("Year"), plt.ylabel("Unemployment Rate")
plt.title("Unemployment Rate 1948-2010")
plt.show()
```

## Treemap

```python
# expenditure for 2008
exp_08 = expenditures[expenditures["year"]==2008]
```

```python
plt.figure(figsize=(14, 14), dpi=80)
squarify.plot(exp_08.expenditure, label=exp_08.category, alpha=.6)
plt.title("Expenditures by Category (2008)")
plt.show()
```

## Stacked Area Chart

```python
test = pd.pivot_table(expenditures, index="year", columns="category", values="expenditure", aggfunc="mean")
```

```python
test
```

```python
test.plot.area()
plt.legend(bbox_to_anchor=(1.05,1.0), loc="upper left")
plt.xlabel("Year"), plt.ylabel("Expenditure")
plt.title("Expenditures 1985-2005")
plt.show()
```
