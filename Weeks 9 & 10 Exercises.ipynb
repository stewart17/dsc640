```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
costco = pd.read_csv("costcos-geocoded.csv")
ppg = pd.read_csv("ppg2008.csv")
```

```python
costco
```

```python
ppg
```

```python
# set index to player names
ppg.set_index("Name  ", inplace=True)
```

```python
ppg.columns
```

## Heat Map

```python
# sort by highest points
ppg.sort_values("PTS", ascending=False, inplace=True)
```

```python
# create heat map
sns.heatmap(ppg, cmap="Blues")
plt.title("Top Scoring NBA Player Stats")
plt.show()
```

## Spatial Chart

```python
# spatial chart of costco locations
fig = px.scatter_geo(costco, lat=costco.Latitude, lon=costco.Longitude)
fig.update_layout(title="Costco Locations in USA", geo_scope="usa")
fig.show()
```

## Contour Chart

```python
# contour chart of points vs Field Goal Attempts
sns.kdeplot(x=ppg.PTS, y=ppg.FGA, shade=True)
plt.title("Relationship Between Field Goal Attempts and Points Scored")
plt.show()
```
